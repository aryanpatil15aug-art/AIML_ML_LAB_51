```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python
data=load_breast_cancer()

x=data.data
y=data.target

print("Classes:",data.target_names)

X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3)
```

```python
lr=LogisticRegression(max_iter=5000)
lr.fit(X_train,Y_train)
pred_lr=lr.predict(X_test)
print("Logistic TestAccuracy:",accuracy_score(Y_test, pred_lr))
pred_lr_train=lr.predict(X_train)
print("Logistic TrainAccuracy:", accuracy_score(Y_train, pred_lr_train))
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
pred_dt=dt.predict(X_test)
print("Decision Tree TestAccuracy:",accuracy_score(Y_test, pred_dt))
pred_dt_train=dt.predict(X_train)
print("Decision Tree TrainAccuracy:", accuracy_score(Y_train, pred_dt_train))
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
pred_knn=knn.predict(X_test)
print("KNeighbors TestAccuracy:",accuracy_score(Y_test, pred_knn))
pred_knn_train=knn.predict(X_train)
print("KNeighbors TrainAccuracy:", accuracy_score(Y_train, pred_knn_train))
```

```python
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
pred_gnb=gnb.predict(X_test)
print("GaussianNB TestAccuracy:",accuracy_score(Y_test, pred_gnb))
pred_gnb_train=gnb.predict(X_train)
print("GaussianNB TrainAccuracy:", accuracy_score(Y_train, pred_gnb_train))
```

```python
mlp=MLPClassifier()
mlp.fit(X_train,Y_train)
pred_mlp= mlp.predict(X_test)
print("MLPClassifier TestAccuracy:",accuracy_score(Y_test, pred_mlp))
pred_mlp_train=mlp.predict(X_train)
print("MLPClassifier TrainAccuracy:", accuracy_score(Y_train, pred_mlp_train))
```

```python
def disp_metrics(Y_test, y_test_pred,algo):
    precision=precision_score(Y_test, y_test_pred)
    recall=recall_score(Y_test,y_test_pred)
    f1=f1_score(Y_test, y_test_pred)
    roc=roc_auc_score(Y_test, y_test_pred)
    
    print("\n============",algo,"=============")
    print("Precision:",precision)
    print("Recall:",recall)
    print("F1-score:",f1)
    print("ROC-AUC:",roc)
```

```python
disp_metrics(Y_test,pred_lr,"Logistic Regression")
disp_metrics(Y_test,pred_dt,"Decision Tree")
disp_metrics(Y_test,pred_knn,"KNeighbors Classifier")
disp_metrics(Y_test,pred_gnb,"GaussianNB Classifier")
disp_metrics(Y_test,pred_mlp,"MLP Classifier")
```

```python
def disp_matrix(Y_test, pred,algo):
    print("\n",algo)
    cm=confusion_matrix(Y_test, pred)
    print("Confusion Matrix:")
    print(cm)
    
    TN2, FP2, FN2, TP2 = cm.ravel()
    type1=FP2/(FP2+TN2)
    type2=FN2/(FN2+TP2)
    
    print("FP (Type-I):",FP2)
    print("FN (Type-II):",FN2)
    print("Type-I Error Rate in %:", type1*100)
    print("Type-II Error Rate in %:", type2*100)
```

```python
disp_matrix(Y_test,pred_lr,"Logistic Regression")
disp_matrix(Y_test,pred_dt,"Decision Tree")
disp_matrix(Y_test,pred_knn,"KNeighbors Classifier")
disp_matrix(Y_test,pred_gnb,"GaussianNB")
disp_matrix(Y_test,pred_mlp,"MLP Classifier")
```
