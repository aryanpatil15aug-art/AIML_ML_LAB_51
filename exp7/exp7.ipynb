```python
import time
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
data=fetch_california_housing()
print(data.keys())
```

```python
x=pd.DataFrame(data.data,columns=data.feature_names)
y=data.target

print(x.head())
print("Samples:", len(x))
print(y[:10])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)
```

```python
def trainer(model, name):
    start=time.perf_counter()
    model.fit(X_train, Y_train)
    model_time=time.perf_counter()-start

    pred_model_train=model.predict(X_train)
    err_model_train=mean_absolute_error(Y_train, pred_model_train)
    
    pred_model_test=model.predict(X_test)
    err_model_test=mean_absolute_error(Y_test, pred_model_test)
    
    print(name," Train MAE:",err_model_train)
    print(name," Test MAE:",err_model_test)
    print(name," Time:",model_time)
    return err_model_test, err_model_train, model_time
```

```python
tree= DecisionTreeRegressor()
err_tree_test,err_tree_train,tree_time=trainer(tree,"Decision Tree")
```

```python
rf= RandomForestRegressor(n_estimators=100)
err_rf_test,err_rf_train,rf_time=trainer(rf,"Random Forest Regressor")
```

```python
gb=GradientBoostingRegressor(n_estimators=200,learning_rate=0.05, max_depth=6)
err_gb_test,err_gb_train,gb_time=trainer(gb,"Gradient Boosting Regressor")
```

```python
print("\n=====Comparison For Train MAC=====")
print("Decision Tree Train:", err_tree_train)
print("Random Forest Train:", err_rf_train)
print("Gradient Boosting Train:", err_gb_train)
```

```python
print("\n=====Comparison For Test MAC=====")
print("Decision Tree Test:", err_tree_test)
print("Random Forest Test:", err_rf_test)
print("Gradient Boosting Test:", err_gb_test)
```

```python
print("\n=====Comparison For Time=====")
print("Decision Tree Time:", tree_time)
print("Random Forest Time:", rf_time)
print("Gradient Boosting Time:",gb_time)
```
